### Stratified Shuffle Split of data

```python
from sklearn.model_selection import train_test_split 

#Split - Stratified shuffle split

def stratified_split_regression(df, column, shareRange, z, random_state):
    shareRange = 850000
    z = 3
    dfSplit = []

    #Splitting dataframe based on ranges
    for i in range(z):
        print(i)
        dfSplit.append(df[df[column].between((shareRange/z)*(i), (shareRange/z)*(i+1))])
        
    #Declaring empty dataframes to append the stratified versions on
    X_train_strat = pd.DataFrame()
    y_train_strat = pd.DataFrame()
    
    X_val_strat = pd.DataFrame()
    y_val_strat = pd.DataFrame()
    
    X_test_strat = pd.DataFrame()
    y_test_strat = pd.DataFrame()
        
    #splitting each range and appending to total areas
    for df in dfSplit:
        y = df[column]
        X = df.loc[:, df.columns != column]

        X_train, X_other, y_train, y_other = train_test_split(X,y,train_size = 0.6,random_state=random_state, shuffle=True)
        X_val, X_test, y_val, y_test = train_test_split(X_other,y_other,train_size = 0.5,random_state=random_state, shuffle=True)    
        
        X_train_strat = pd.concat([X_train_strat, X_train], ignore_index=True, sort=False)
        y_train_strat = pd.concat([y_train_strat, y_train], ignore_index=True, sort=False)

        X_val_strat = pd.concat([X_val_strat, X_val], ignore_index=True, sort=False)
        y_val_strat = pd.concat([y_val_strat, y_val], ignore_index=True, sort=False)

        X_test_strat = pd.concat([X_test_strat, X_test], ignore_index=True, sort=False)
        y_test_strat = pd.concat([y_test_strat, y_test], ignore_index=True, sort=False)
    
    return X_train_strat, y_train_strat, X_val_strat, y_val_strat, X_test_strat, y_test_strat
    
X_train, y_train, X_val, y_val, X_test, y_test = stratified_split_regression(df, ' shares', 850000, 3, 42)

print(y_train)
print(y_val)
print(y_test)
```
